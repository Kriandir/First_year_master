## Question 1

```python
diceroll = np.genfromtxt("dicerolls.txt", dtype=np.float, skip_header=1, autostrip=True, usecols=(0,1,2))
dice = {}
tstattest = []
pvaltest = []
tmax= []
tmin = []
pvalues = []
for i in range (1,101):
    dice[i] =[]

for s in diceroll:
    dice[s[0]].append(s[2])

for i in range(1,len(dice)+1):
    tstat, pval = ss.ttest_1samp(dice[i], popmean=14)
    tstattest.append(tstat)
    pvaltest.append(pval)

for i in range(len(tstattest)):
    if tstattest[i] == max(tstattest):
        tmax.append([tstattest[i],i+1])
    if tstattest[i] == min(tstattest):
        tmin.append([tstattest[i],i+1])
    
for i in range(len(pvaltest)):
    if pvaltest[i] <0.01:
        pvalues.append([pvaltest[i],i+1])
dicelist=[]

for i in range(1,101,1):
    dicelist.append(np.mean(dice[i]))

for i in pvalues:
 
    if i[1] == tmax[0][1]:
        print "We have a maximum value with tstat of %0.3f and a pvalue of %0.4f at dice %i with a mean of %.2f" \
        %(tmax[0][0],i[0],i[1],np.mean(dice[i[1]]))
    if i[1] == tmin[0][1]:
        print "We have a minimum value with tstat of %0.3f and a pvalue of %0.4f at dice %i with a mean of %.2f" \
        %(tmin[0][0],i[0],i[1],np.mean(dice[i[1]]))

print 'So going from the pvalues we found we have dice 64 deviating from the mean postively and dice 94 deviating from the mean\
negatively.\n However since these pvalues come from only a hunderd trials we have to correct for the probability that no outlier occurs\
which is (1-pvalue)^ntrails which in the case of our positively deviating dice is: %0.3f and in \
the case of our negatively deviating dice is %0.3f which means that our positvely deviating dice\
has a 14.5 percent chance of being a coincidence and our negatively deviating dice has\
a 28 percent chance of being a coincidence. Based on this we can not sufficiently conclude if the dice are loaded or not' %(((1-pvalues[0][0])**100),(1-pvalues[1][0])**100)
fig = plt.figure(figsize=[15,15])
plt.plot(np.arange(0,100,1),dicelist)
plt.axhline(14, color='r', linestyle='dotted', lw=2)
plt.axis([-1,101,12.5,15.5])
plt.xlabel("Dice",fontsize=18)
plt.ylabel("Mean",fontsize=18)
plt.show()
```

```python
def p_to_sigmas(pval):
    nd = ss.norm(0., 1.)
# We use the 'inverse survival function', the inverse of the cdf. We also need to divide
# our p-value by 2 to account for the negative side of the normal distribution.
    return nd.isf(pval/2.)

lamb = 2.3
p = ss.poisson(lamb)
# get probability (pvalue and sigma)for exactly 7 events with a poisson distribution frozen with expected rate 2.3
a = p.sf(6)
b = p_to_sigmas(a)

print "The p-value found is %0.3f and the sigma found is %0.3f this means that we can't really confidently state \nthat what we measure is not background. \
For such a statement to happen we would need a sigma of atleast 3" %(a,b)
```

b/

```python
g = -1
d = 1

while p_to_sigmas(d) < 3:
    d=p.sf(g)
    g+=1
    

print "We find for an amount of events of %i that the p-value found is %0.3f and the sigma is %0.3f which is \nabove the 3 sigma upperlimit and thus significant" % (g,d,p_to_sigmas(d))
```

## Question 4

```python
star = np.genfromtxt ("star.txt", dtype=np.float, names=["time", "flux1","err1","flux2","err2"], skip_header=1, autostrip=True)

time = star["time"]
star1 = [star["flux1"],star["err1"]]
star2 = [star["flux2"],star["err2"]]
# plot star1
fig, (ax1,ax2) = plt.subplots(1,2,figsize=[20,10])
ax1.errorbar(time,star1[0],star1[1],fmt='o')
ax1.set_xlabel("Time in days",fontsize=18)
ax1.set_ylabel("Flux",fontsize=18)
ax2.set_xlabel("Time in days",fontsize=18)
ax2.set_ylabel("Residuals",fontsize=18)

# since the error's are normally distributed i can use curve fit so..
xval = time
yval = star1[0]
dy = star1[1]
### We define a new model function here since curve_fit requires parameters to be listed separately
def my_cfmodel(value, a): 
    return a

def ChiSq(parm, xval, yval, dy): # the weighted least-squares
    ymod = my_cfmodel(xval, parm)
    chisq = sum(pow((yval-ymod)/dy,2))
    return chisq

p0 = [16]  # Define starting values
ml_cfpars, ml_cfcovar = op.curve_fit(my_cfmodel, xval, yval, p0, sigma=dy)
err = np.sqrt(np.diag(ml_cfcovar))

print "a = %0.2f +/- %0.2f thus a confidence interval of 1 sigma between %0.2f and %0.2f" % (ml_cfpars[0],err[0],ml_cfpars[0]-err[0],ml_cfpars[0]+err[0])


ax1.axhline(my_cfmodel(xval,ml_cfpars[0]),color= "red",linestyle="dotted",lw=3)
ax2.errorbar(xval, yval - my_cfmodel(xval,ml_cfpars[0]), yerr=dy, marker="o", linestyle='')
ax2.plot(xval, yval - my_cfmodel(xval,ml_cfpars[0]), 'o')
ax2.axhline(0.0, color='r', linestyle='dotted', lw=2)
dof = len(xval)-len(p0)
chisquare = ChiSq(ml_cfpars,xval,yval,dy)
### curve_fit does not return the minimum chi-squared so we must calculate that ourselves for the MLEs
### obtained by the fit, e.g. using our original ChiSq function 
print "Minimum Chi-squared = %0.4f for %i d.o.f" % (chisquare,dof)
print "With a goodness-of-fit of %0.4f" % (ss.chi2.sf(chisquare, df=dof))
print "Since the goodness-of-fit is low but not totally insignificant we can't clearly say that the star is variable or not"
```

```python
def histrebin(mininbin,counts,edges): # rebins a histogram so there are at least mininbin counts per bin

    edges2=np.zeros(1)
    counts2=np.zeros(1)
    i=0
    countsum=0
    firstbin_flag = 1
    edges2[0]=edges[0]
    for x in counts:
        countsum=countsum+x
        i=i+1
        if countsum >= mininbin and (sum(counts[i:]) >= mininbin or i == len(counts)):  # The sum over counts in 
            # remaining bins ensures that we are not left with a bin with counts less than our limit at the end.
            if firstbin_flag == 1: # if this is the first bin satisfying our requirement, 
                # do not append but replace bin 0 with countsum
                counts2[0]=countsum
                firstbin_flag = 0 # now we read the first bin we can switch this off
            else:
                counts2=np.append(counts2,countsum)
            edges2=np.append(edges2,edges[i])
            countsum=0        

    return counts2, edges2


def my_fit_model(xval, parm):
    return ss.lognorm.cdf(xval, s=parm[1], loc=parm[0], scale=parm[2])*len(realrate)

def ChiSqBin(parm, xval, y, dy): # the weighted least-squares
    mod = my_fit_model(xval, parm)
    
    ymod1 = np.diff(mod)
    
    k = 0
    ymod = np.zeros(len(y))
    
    for k in range(len(xval) - 1):
        ymod[k] = ymod1[k] / (xval[k+1] - xval[k])
        k+=1

    return sum(pow((y-ymod)/dy,2))


# Load in text erase zeroes
star = np.genfromtxt ("cygx1_xray.dat", dtype=np.float, names=["bin","rate"], skip_header=1, autostrip=True)

rate = star["rate"]
realrate=[]
for i in range(len(rate)):
    if rate[i] != 0.0:
        realrate.append(rate[i]/1000)


#         Get histogram values

counts, edges = np.histogram(realrate, bins=50, range=[min(realrate),max(realrate)], density=False)

#  rebin everything

counts2,edges2 = histrebin(20,counts,edges)

# get edges
x=[]
for i in range(len(edges2)-1):
    x.append((edges2[i+1]+edges2[i])/2)

bwidths = np.diff(edges2) # calculates the width of each bin
y = counts2/bwidths # determines the count densities
yerr = np.sqrt(counts2)/bwidths # calculate the error
xerr = bwidths/2

fig,(ax1,ax2) = plt.subplots(2,1,figsize=[20,20])
ax1.errorbar(x,y,xerr=xerr,yerr=yerr,fmt='o')


# guess values
mu = 0.16
sigm = 0.3
scale = 0.9

fparm = [mu,sigm,scale]
ml_pars, ml_funcval, ml_grad, ml_covar, func_calls, grad_calls, warnflag = \
op.fmin_bfgs(ChiSqBin, fparm, args=(edges2, y, yerr), full_output=True)
err = np.sqrt(np.diag(ml_covar))

# see how good the model fits
modelfit = ss.lognorm.pdf(x, loc=ml_pars[0], s=ml_pars[1], scale=ml_pars[2])*len(realrate)


print "mu = %0.3f +/- %0.3f and thus a confidence interval of 1 sigma between %0.3f and %0.3f" % (ml_pars[0], err[0],ml_pars[0]-err[0],ml_pars[0]+err[0])
print "sigma = %0.4f +/- %0.4f and thus a confidence interval of 1 sigma between %0.4f and %0.4f" % (ml_pars[1], err[1],ml_pars[1]-err[1],ml_pars[1]+err[1])
print "scale = %0.3f +/- %0.3f and thus a confidence interval of 1 sigma between %0.3f and %0.3f" % (ml_pars[2], err[2],ml_pars[2]-err[2],ml_pars[2]+err[2])
print "Minimum Chi-squared = %0.3f with %i degrees of freedom" %(ml_funcval,len(y)-len(fparm))
print "The goodness of fit is: %0.3f"  % ss.chi2.sf(ml_funcval,df=(len(y)-len(fparm)))
print "This fit is very significant with a goodness of fit value of 0.446 and also as can be seen in the plot follows the data distribution quite well \
this makes a lognormal fit model a very good model for this kind of data"

ax1.plot(x,modelfit,color= "red",linestyle="dotted",lw=3)

ax2.errorbar(x, y - modelfit, xerr=xerr, yerr=yerr, fmt='o')
ax2.plot(x, y - modelfit, 'o')
ax2.axhline(0.0, color='r', linestyle='dotted', lw=2)

ax2.set_xlabel("Normalize photon rate", fontsize=18)
ax2.set_ylabel("Residuals", fontsize=18)
ax1.set_xlabel("Normalized photon rate", fontsize=18)
ax1.set_ylabel("Counts per bin", fontsize=18)
plt.show()
```

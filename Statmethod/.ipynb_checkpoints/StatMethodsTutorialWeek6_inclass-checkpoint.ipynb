## Fitting with parameter constraints 

It is often useful to be able to fit models to data while constraining one or more of the model parameters.  If you already have a good idea what a parameter should be (or should be close to), then applying such constraints can help the optimiser to find the maximum likelihood for the model more quickly, or the fit be prevented from getting stuck in a local maximum/minimum or from wandering to some extreme parameter values.  Furthermore, being able to constrain parameters (and conversely, to allow them to vary) is crucial to doing hypothesis testing where we compare <i>nested models</i>.

We will first take a look at some approaches to constraining parameters when doing model-fitting using `scipy.optimize`.  First we will set up our useful functions.  Note in particular that we have now made our likelihood functions able to cope with generic models, by specifying the model function name as one of the parameters of the likelihood function:

```python
### Import the scipy optimisation package
import scipy.optimize as op

###
### The two forms of the likelihood - note that the function LogLikelihood assumes normally distributed errors
### These functions also read in the name of the model as a parameter my_model, so they can be used generically
### without having to always used the same model name.
###

def LogLikelihood(parm, my_model, xval, yval, dy): # the -ve log-likelihood
    ymod = my_model(xval, *parm)  #We define our 'physical model' separately
    nd = scipy.stats.norm(ymod, dy) #we define our normal distribution
    return -sum(np.log(nd.pdf(yval)))

def ChiSq(parm, my_model, xval, yval, dy): # the weighted least-squares
    ymod = my_model(xval, *parm)
    chisq = sum(pow((yval-ymod)/dy,2))
    return chisq

###
### And to convert significances from sigma to p and vice versa
###

def sigmas_to_p(sigval):
    nd = scipy.stats.norm(0., 1.)
    return 2.*nd.sf(sigval)

def p_to_sigmas(pval):
    nd = scipy.stats.norm(0., 1.)
# We use the 'inverse survival function', the inverse of the sf. We also need to divide
# our p-value by 2 to account for the negative side of the normal distribution.
    return nd.isf(pval/2.)
```

We will learn about parameter constraints using Reynolds' fluid-flow data.  In order to see how parameter constraints can help us to fit more complex models, this time we will read in the full data.  We will plot it first to remind ourselves what it looks like:

```python
reynolds = np.genfromtxt ("reynolds.txt", dtype=np.float, names=["dP", "v"], skip_header=1, autostrip=True)
## change units
ppm = 9.80665e3
dp = reynolds["dP"]*ppm
v = reynolds["v"]
dv = 6.3e-3  ## We will assume a constant error for the purposes of this exercise
### Let's plot the data
fig, ax1 = plt.subplots(1,1, figsize=(6,6))
ax1.errorbar(dp, v, yerr=dv, marker="o", linestyle="")
ax1.set_xlabel("Pressure gradient (Pa/m)", fontsize=20)
ax1.set_ylabel("Velocity (m/s)", fontsize=20)
ax1.tick_params(labelsize=16)
plt.show()
```

### Parameter constraints with `curve_fit`

As a test-run, let's first fit the data with a simple linear model which we used to successfully fit the first 9 data points in the previous weeks.  We will start by using `curve_fit` to fit the data.  This is appropriate since we assume that the errors are normally distributed (remember that `curve_fit` uses weighted-least-squares fitting and so implicitly makes this assumption):

```python
def lin_mod(value, a, b):
    return a + b*value

my_model = lin_mod # Now we assign the name of the model to my_model, which sets the input for the optimisation and
# likelihood functions.

p0 = [0.01, 0.0035]  # Define starting values
ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(my_model, dp, v, p0, sigma=dv)
err = np.sqrt(np.diag(ml_cfcovar))

print "a = " + str(ml_cfpars[0]) + " +/- " + str(err[0])
print "b = " + str(ml_cfpars[1]) + " +/- " + str(err[1])

### curve_fit does not return the minimum chi-squared so we must calculate that ourselves for the MLEs
### obtained by the fit, e.g. using our original ChiSq function 
print "Minimum Chi-squared = " + str(ChiSq(ml_cfpars,my_model,dp,v,dv)) + " for " + str(len(dp)-len(p0)) + " d.o.f."
```

Not surprisingly, the fit isn't very good!  We could plot the best-fitting linear model against the data to show the problem, but it should already be obvious from the plot of the data above that a linear model will not work to explain the full data-set.  Physically, turbulence sets in above a certain pressure-gradient, and this effect manifests itself as a sharp bend or 'break' in the linear model.  To fit the data better we will consider a more complex model consisting of two linear slopes which join at a 'break' value for the pressure gradient.  Note that this is still a purely empirical model which we choose for demonstration purposes only.  A more physically-motivated fit might suppose some model for the turbulence that enables the shape to be modelled fully self-consistently. 

```python
def bklin_mod(value, a, b1, b2, bk):  ###  A 'broken' linear model with two gradients (b1 and b2)
    ### The lines join at a break value (bk).  The intercept on the y-axis is still denoted a.
    y = np.empty(len(value))
    i = 0
    while (i < len(value)): 
        if value[i] < bk:
            y[i]= a + b1*value[i]
        else:
            y[i] = a + b1*bk + b2*(value[i]-bk)  # Note that we have to ensure that the lines join at value = bk!
        i += 1
    return y

my_model = bklin_mod
p0 = [0.01, 0.0035, 0.0035, 60.0] # Try some starting parameters

ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(my_model, dp, v, p0, sigma=dv)
err = np.sqrt(np.diag(ml_cfcovar))
print "a = " + str(ml_cfpars[0]) + " +/- " + str(err[0])
print "b1 = " + str(ml_cfpars[1]) + " +/- " + str(err[1])
print "b2 = " + str(ml_cfpars[2]) + " +/- " + str(err[2])
print "Break value = " + str(ml_cfpars[3]) + " +/- " + str(err[3])
print "Minimum Chi-squared = " + str(ChiSq(ml_cfpars,my_model,dp,v,dv)) + " for " + str(len(dp)-len(p0)) + " d.o.f."
```

The fit is still bad!  But note that the model parameters for `a` and `b1` have converged on the same values as the intercept and gradient of the simple linear fit.  This is because the break has wandered off to an extremely high value (much larger than the observed range of `dp`), so our broken linear model effectively becomes the same as the simple linear model. The covariance matrix returns infinities for the errors as a result of this problem.

In order to help the fit along, let's constrain the break parameter so that it cannot wander off.  We'll set it to a best-guess value of 60 Pa/m.  In order to constrain the `bk` parameter in `curve_fit`, we will use a useful Python trick of substituting our broken linear model function with a simpler version where bk is not one of the variable parameters.  We do this using Python's special `lambda` anonymous function:

```python
p0 = [0.01, 0.0035, 0.0035]
#ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(my_model, dp, v, p0, sigma=dv)
bk = 60.0 
ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(lambda x, a, b1, b2: my_model(x, a, b1, b2, bk), dp, v, p0, sigma=dv)
## lambda is used as the function to fit to the data.  It only takes x and the three variables listed, so bk takes 
## whatever value we choose.
err = np.sqrt(np.diag(ml_cfcovar))
print "a = " + str(ml_cfpars[0]) + " +/- " + str(err[0])
print "b1 = " + str(ml_cfpars[1]) + " +/- " + str(err[1])
print "b2 = " + str(ml_cfpars[2]) + " +/- " + str(err[2])
print "Break value = ", bk
bestpars = [ml_cfpars[0], ml_cfpars[1], ml_cfpars[2], bk] # We have to add bk back in to our parameters to use
# our ChiSq function.
print "Minimum Chi-squared = " + str(ChiSq(bestpars,my_model,dp,v,dv)) + " for " + str(len(dp)-len(p0)) + " d.o.f."
```

Now we see that we obtain a much better fit!  The fit is still not optimal, but at least the parameters have already converged to be closer to the true values, so there is much less risk that the fit will go wrong if we now allow `bk` to also be a fitted (i.e. 'free') parameter.  We will therefore finish the fit and also plot the best-fitting model against the data: 

```python
p0 = [ml_cfpars[0], ml_cfpars[1], ml_cfpars[2], 60.0]
ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(my_model, dp, v, p0, sigma=dv)
err = np.sqrt(np.diag(ml_cfcovar))
print "a = " + str(ml_cfpars[0]) + " +/- " + str(err[0])
print "b1 = " + str(ml_cfpars[1]) + " +/- " + str(err[1])
print "b2 = " + str(ml_cfpars[2]) + " +/- " + str(err[2])
print "Break value = " + str(ml_cfpars[3]) + " +/- " + str(err[3])
print "Minimum Chi-squared = " + str(ChiSq(ml_cfpars,my_model,dp,v,dv)) + " for " + str(len(dp)-len(p0)) + " d.o.f."

fig, ax1 = plt.subplots(1,1, figsize=(6,6))
x = np.arange(0.0, 120.0, 1.0)
ax1.plot(x, my_model(x, *ml_cfpars))
ax1.errorbar(dp, v, yerr=dv, marker="o", linestyle="")
ax1.set_xlabel("Pressure gradient (Pa/m)", fontsize=20)
ax1.set_ylabel("Velocity (m/s)", fontsize=20)
ax1.tick_params(labelsize=16)
plt.show()
```

### Other methods to constrain parameters

Depending on the optimisation function used, a number of other approaches can be used to provide parameter constraints, e.g. for many functions we can specify 'bounds' (i.e. limiting ranges) on the parameter values.  This is possible for `curve_fit`, but here we show another example using the `L-BFGS-B` method with the `minimize` function:

```python
p0 = [0.01, 0.0035, 0.0035, 60.0]

result =  op.minimize(ChiSq, p0, args=(my_model,dp,v,dv), method='L-BFGS-B', 
                     bounds=[(0.0,None),(0.0,None),(0.0,None),(60.0,60.0)])
print result.x # best-fitting parameter values
print result.fun # value of the function minimised (e.g. chi-squared)
```

We can also use the bounds to constrain the parameter space searched so that the best fit can be found without the fit getting stuck, e.g.:

```python
p0 = [0.01, 0.0035, 0.0035, 60.0]

result =  op.minimize(ChiSq, p0, args=(my_model,dp,v,dv), method='L-BFGS-B', 
                     bounds=[(0.0,None),(0.0,None),(0.0,None),(0.0,120.0)])
print result.x # best-fitting parameter values
print result.fun # value of the function minimised (e.g. chi-squared)
```

You can see that the best-fitting parameters are close to those obtained from curve_fit.  

### Using constraints to obtain 1-dimensional confidence intervals with brute force search

Now we will look at how to use constraints to find 1-dimensional confidence intervals on a parameter (i.e. marginalised over the other parameters) using brute force search:

```python
bestbk = ml_cfpars[3]
bk = 50
i=50
stepsize = 0.01
confidence=[]
ylist= []
xlist=[]

p0=[ml_cfpars[0],ml_cfpars[1],ml_cfpars[2],bk]
bestchisq = ChiSq(ml_cfpars,my_model,dp,v,dv)
while i < 75:
    p0.pop()
    p0.append(bk)
    result = op.minimize(ChiSq,p0,args=(my_model,dp,v,dv),method='L-BFGS-B',\
                             bounds=[(0.0,None),(0.0,None),(0.0,None),(bk,bk)])
    ylist.append(result.fun-bestchisq)
    i+=stepsize
    bk=i
    xlist.append(i)
plt.plot(xlist,ylist)
for i in range(len(ylist)):
    if ylist[i] == min(ylist):
        foundvalue = xlist[i]
        
    if  (1)<ylist[i]<(1+stepsize):
        print xlist[i]
        confidence.append(xlist[i])
print result
        
plt.show()
print "found value of %0.3f with a confidence interval of %0.3f and +%0.3f" %(foundvalue,confidence[0]-foundvalue,confidence[1]-foundvalue)
```

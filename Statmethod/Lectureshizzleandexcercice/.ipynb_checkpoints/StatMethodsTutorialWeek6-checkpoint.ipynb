## Fitting with parameter constraints 

It is often useful to be able to fit models to data while constraining one or more of the model parameters.  If you already have a good idea what a parameter should be (or should be close to), then applying such constraints can help the optimiser to find the maximum likelihood for the model more quickly, or the fit be prevented from getting stuck in a local maximum/minimum or from wandering to some extreme parameter values.  Furthermore, being able to constrain parameters (and conversely, to allow them to vary) is crucial to doing hypothesis testing where we compare <i>nested models</i>.

We will first take a look at some approaches to constraining parameters when doing model-fitting using `scipy.optimize`.  First we will set up our useful functions.  Note in particular that we have now made our likelihood functions able to cope with generic models, by specifying the model function name as one of the parameters of the likelihood function:

```python
### Import the scipy optimisation package
import scipy.optimize as op

###
### The two forms of the likelihood - note that the function LogLikelihood assumes normally distributed errors
### These functions also read in the name of the model as a parameter my_model, so they can be used generically
### without having to always used the same model name.
###

def LogLikelihood(parm, my_model, xval, yval, dy): # the -ve log-likelihood
    ymod = my_model(xval, *parm)  #We define our 'physical model' separately
    nd = scipy.stats.norm(ymod, dy) #we define our normal distribution
    return -sum(np.log(nd.pdf(yval)))

def ChiSq(parm, my_model, xval, yval, dy): # the weighted least-squares
    ymod = my_model(xval, *parm)
    chisq = sum(pow((yval-ymod)/dy,2))
    return chisq

###
### And to convert significances from sigma to p and vice versa
###

def sigmas_to_p(sigval):
    nd = scipy.stats.norm(0., 1.)
    return 2.*nd.sf(sigval)

def p_to_sigmas(pval):
    nd = scipy.stats.norm(0., 1.)
# We use the 'inverse survival function', the inverse of the sf. We also need to divide
# our p-value by 2 to account for the negative side of the normal distribution.
    return nd.isf(pval/2.)
```

We will learn about parameter constraints using Reynolds' fluid-flow data.  In order to see how parameter constraints can help us to fit more complex models, this time we will read in the full data.  We will plot it first to remind ourselves what it looks like:

```python
reynolds = np.genfromtxt ("reynolds.txt", dtype=np.float, names=["dP", "v"], skip_header=1, autostrip=True)
## change units
ppm = 9.80665e3
dp = reynolds["dP"]*ppm
v = reynolds["v"]
dv = 6.3e-3  ## We will assume a constant error for the purposes of this exercise
### Let's plot the data
fig, ax1 = plt.subplots(1,1, figsize=(6,6))
ax1.errorbar(dp, v, yerr=dv, marker="o", linestyle="")
ax1.set_xlabel("Pressure gradient (Pa/m)", fontsize=20)
ax1.set_ylabel("Velocity (m/s)", fontsize=20)
ax1.tick_params(labelsize=16)
plt.show()
```

### Parameter constraints with `curve_fit`

As a test-run, let's first fit the data with a simple linear model which we used to successfully fit the first 9 data points in the previous weeks.  We will start by using `curve_fit` to fit the data.  This is appropriate since we assume that the errors are normally distributed (remember that `curve_fit` uses weighted-least-squares fitting and so implicitly makes this assumption):

```python
def lin_mod(value, a, b):
    return a + b*value

my_model = lin_mod # Now we assign the name of the model to my_model, which sets the input for the optimisation and
# likelihood functions.

p0 = [0.01, 0.0035]  # Define starting values
ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(my_model, dp, v, p0, sigma=dv)
err = np.sqrt(np.diag(ml_cfcovar))

print "a = " + str(ml_cfpars[0]) + " +/- " + str(err[0])
print "b = " + str(ml_cfpars[1]) + " +/- " + str(err[1])

### curve_fit does not return the minimum chi-squared so we must calculate that ourselves for the MLEs
### obtained by the fit, e.g. using our original ChiSq function 
print "Minimum Chi-squared = " + str(ChiSq(ml_cfpars,my_model,dp,v,dv)) + " for " + str(len(dp)-len(p0)) + " d.o.f."
```

Not surprisingly, the fit isn't very good!  We could plot the best-fitting linear model against the data to show the problem, but it should already be obvious from the plot of the data above that a linear model will not work to explain the full data-set.  Physically, turbulence sets in above a certain pressure-gradient, and this effect manifests itself as a sharp bend or 'break' in the linear model.  To fit the data better we will consider a more complex model consisting of two linear slopes which join at a 'break' value for the pressure gradient.  Note that this is still a purely empirical model which we choose for demonstration purposes only.  A more physically-motivated fit might suppose some model for the turbulence that enables the shape to be modelled fully self-consistently. 

```python
def bklin_mod(value, a, b1, b2, bk):  ###  A 'broken' linear model with two gradients (b1 and b2)
    ### The lines join at a break value (bk).  The intercept on the y-axis is still denoted a.
    y = np.empty(len(value))
    i = 0
    while (i < len(value)): 
        if value[i] < bk:
            y[i]= a + b1*value[i]
        else:
            y[i] = a + b1*bk + b2*(value[i]-bk)  # Note that we have to ensure that the lines join at value = bk!
        i += 1
    return y

my_model = bklin_mod
p0 = [0.01, 0.0035, 0.0035, 60.0] # Try some starting parameters

ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(my_model, dp, v, p0, sigma=dv)
err = np.sqrt(np.diag(ml_cfcovar))
print "a = " + str(ml_cfpars[0]) + " +/- " + str(err[0])
print "b1 = " + str(ml_cfpars[1]) + " +/- " + str(err[1])
print "b2 = " + str(ml_cfpars[2]) + " +/- " + str(err[2])
print "Break value = " + str(ml_cfpars[3]) + " +/- " + str(err[3])
print "Minimum Chi-squared = " + str(ChiSq(ml_cfpars,my_model,dp,v,dv)) + " for " + str(len(dp)-len(p0)) + " d.o.f."
```

The fit is still bad!  But note that the model parameters for `a` and `b1` have converged on the same values as the intercept and gradient of the simple linear fit.  This is because the break has wandered off to an extremely high value (much larger than the observed range of `dp`), so our broken linear model effectively becomes the same as the simple linear model. The covariance matrix returns infinities for the errors as a result of this problem.

In order to help the fit along, let's constrain the break parameter so that it cannot wander off.  We'll set it to a best-guess value of 60 Pa/m.  In order to constrain the `bk` parameter in `curve_fit`, we will use a useful Python trick of substituting our broken linear model function with a simpler version where bk is not one of the variable parameters.  We do this using Python's special `lambda` anonymous function:

```python
p0 = [0.01, 0.0035, 0.0035]
#ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(my_model, dp, v, p0, sigma=dv)
bk = 60.0 
ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(lambda x, a, b1, b2: my_model(x, a, b1, b2, bk), dp, v, p0, sigma=dv)
## lambda is used as the function to fit to the data.  It only takes x and the three variables listed, so bk takes 
## whatever value we choose.
err = np.sqrt(np.diag(ml_cfcovar))
print "1 = " + str(ml_cfpars[0]) + " +/- " + str(err[0])
print "b1 = " + str(ml_cfpars[1]) + " +/- " + str(err[1])
print "b2 = " + str(ml_cfpars[2]) + " +/- " + str(err[2])
print "Break value = ", bk
bestpars = [ml_cfpars[0], ml_cfpars[1], ml_cfpars[2], bk] # We have to add bk back in to our parameters to use
# our ChiSq function.
print "Minimum Chi-squared = " + str(ChiSq(bestpars,my_model,dp,v,dv)) + " for " + str(len(dp)-len(p0)) + " d.o.f."
```

Now we see that we obtain a much better fit!  The fit is still not optimal, but at least the parameters have already converged to be closer to the true values, so there is much less risk that the fit will go wrong if we now allow `bk` to also be a fitted (i.e. 'free') parameter.  We will therefore finish the fit and also plot the best-fitting model against the data: 

```python
p0 = [ml_cfpars[0], ml_cfpars[1], ml_cfpars[2], 60.0]
ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(my_model, dp, v, p0, sigma=dv)
err = np.sqrt(np.diag(ml_cfcovar))
print "a = " + str(ml_cfpars[0]) + " +/- " + str(err[0])
print "b1 = " + str(ml_cfpars[1]) + " +/- " + str(err[1])
print "b2 = " + str(ml_cfpars[2]) + " +/- " + str(err[2])
print "Break value = " + str(ml_cfpars[3]) + " +/- " + str(err[3])
print "Minimum Chi-squared = " + str(ChiSq(ml_cfpars,my_model,dp,v,dv)) + " for " + str(len(dp)-len(p0)) + " d.o.f."

fig, ax1 = plt.subplots(1,1, figsize=(6,6))
x = np.arange(0.0, 120.0, 1.0)
ax1.plot(x, my_model(x, *ml_cfpars))
ax1.errorbar(dp, v, yerr=dv, marker="o", linestyle="")
ax1.set_xlabel("Pressure gradient (Pa/m)", fontsize=20)
ax1.set_ylabel("Velocity (m/s)", fontsize=20)
ax1.tick_params(labelsize=16)
plt.show()
```

### Other methods to constrain parameters

Depending on the optimisation function used, a number of other approaches can be used to provide parameter constraints, e.g. for many functions we can specify 'bounds' (i.e. limiting ranges) on the parameter values.  This is possible for `curve_fit`, but here we show another example using the `L-BFGS-B` method with the `minimize` function:

```python
p0 = [0.01, 0.0035, 0.0035, 60.0]

result =  op.minimize(ChiSq, p0, args=(my_model,dp,v,dv), method='L-BFGS-B', 
                     bounds=[(0.0,None),(0.0,None),(0.0,None),(60.0,60.0)])
print result.x # best-fitting parameter values
print result.fun # value of the function minimised (e.g. chi-squared)
```

We can also use the bounds to constrain the parameter space searched so that the best fit can be found without the fit getting stuck, e.g.:

```python
p0 = [0.01, 0.0035, 0.0035, 60.0]

result =  op.minimize(ChiSq, p0, args=(my_model,dp,v,dv), method='L-BFGS-B', 
                     bounds=[(0.0,None),(0.0,None),(0.0,None),(0.0,120.0)])
print result.x # best-fitting parameter values
print result.fun # value of the function minimised (e.g. chi-squared)
```

You can see that the best-fitting parameters are close to those obtained from curve_fit.  

### Using constraints to obtain 1-dimensional confidence intervals with brute force search

Now we will look at how to use constraints to find 1-dimensional confidence intervals on a parameter (i.e. marginalised over the other parameters) using brute force search.  For convenience we will first record the best-fitting parameters and chi-squared.  We then step through a range of values for the break, and plot the resulting $\Delta \chi^{2}$.  The 1-$\sigma$ confidence interval corresponds to $\Delta \chi^{2} = 1.0$.  We also use a trick to obtain the minimum and maximum value in the vector of chi-squared values for which $\Delta \chi^{2} \leq 1.0$, which is a way to return the confidence intervals by finding the parameter values corresponding to those estimates.

```python
bestbk = ml_cfpars[3]
bestchisq = ChiSq(ml_cfpars,my_model,dp,v,dv)

p0 = [0.01, 0.0035, 0.0035]
#ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(my_model, dp, v, p0, sigma=dv)
bkvals = np.arange(50.0, 70.0, 0.01)
chisqval = []
for bk in bkvals: 
    ml_cfpars, ml_cfcovar = scipy.optimize.curve_fit(lambda x, a, b1, b2: my_model(x, a, b1, b2, bk), dp, v, p0, 
                                                     sigma=dv)
    bestpars = [ml_cfpars[0], ml_cfpars[1], ml_cfpars[2], bk]
    chisqval = np.append(chisqval, ChiSq(bestpars,my_model,dp,v,dv)-bestchisq)
    
fig, ax1 = plt.subplots(1,1, figsize=(6,6))
x = np.arange(0.0, 120.0, 1.0)
ax1.plot(bkvals, chisqval)
ax1.set_xlabel("Break (Pa/m)", fontsize=20)
ax1.set_ylabel("delta-chi^2", fontsize=20)
ax1.tick_params(labelsize=16)
plt.show()

minbk = np.min([bk for bk, xsq in zip(bkvals, chisqval) if xsq <= 1.0])
maxbk = np.max([bk for bk, xsq in zip(bkvals, chisqval) if xsq <= 1.0])

print "Break is " + str(bestbk) + " - " + str(bestbk-minbk) + " + " + str(maxbk-bestbk)
```

Or for presentation purposes: The break value is $62.6^{+2.7}_{-2.5}$ Pa/m

##  Application to hypothesis testing: comparing models and freeing up parameters

We will now look at how to use constraints to carry out hypothesis testing to compare nested models fitted to data.  Let's consider the simulated photon data which we first looked at in Week 5, which can be well-fitted with a power-law distribution.  We will first load and plot the data, we can also rebin at this stage so that there are at least 20 counts per bin and hence the errors on each bin can be assumed to be normally distributed.

```python
### First we will define our histogram rebinning function
def histrebin(mininbin,counts,edges): # rebins a histogram so there are at least mininbin counts per bin

    edges2=np.zeros(1)
    counts2=np.zeros(1)
    i=0
    countsum=0
    firstbin_flag = 1
    edges2[0]=edges[0]
    for x in counts:
        countsum=countsum+x
        i=i+1
        if countsum >= mininbin and (sum(counts[i:]) >= mininbin or i == len(counts)):  # The sum over counts in 
            # remaining bins ensures that we are not left with a bin with counts less than our limit at the end.
            if firstbin_flag == 1: # if this is the first bin satisfying our requirement, 
                # do not append but replace bin 0 with countsum
                counts2[0]=countsum
                firstbin_flag = 0 # now we read the first bin we can switch this off
            else:
                counts2=np.append(counts2,countsum)
            edges2=np.append(edges2,edges[i])
            countsum=0        

    return counts2, edges2

###  First read in the data.  This is a simple (single-column) list of energies:
photens = np.genfromtxt('photon_energies.txt')
emin=10.0
emax=200.0

### Now we make our unbinned histogram.  We can keep the initial number of bins relatively large.
nbins = 50
counts, edges = np.histogram(photens, bins=nbins, range=[emin,emax], density=False)

### And now we use our new function to rebin so there are at least mincounts counts per bin:
mincounts = 20  # Here we set it to our minimum requirement of 20, but in principle you could set it higher
counts2, edges2 = histrebin(mincounts,counts,edges)

bwidths = np.diff(edges2) # calculates the width of each bin
cdens = counts2/bwidths # determines the count densities
cdens_err = np.sqrt(counts2)/bwidths # calculate the errors: remember the error is based on the counts, 
# not the count density, so we have to also apply the same normalisation.

energies = (edges2[:-1]+edges2[1:])/2.  # This calculates the energy bin centres

fig, ax1 = plt.subplots(1,1, figsize=(6,6))
ax1.errorbar(energies, cdens, xerr=bwidths/2., yerr=cdens_err, fmt='o')
ax1.set_xlabel("Energy (GeV)", fontsize=20)
ax1.set_ylabel("Counts/GeV", fontsize=20)
ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_ylim(0.3,300.0)
ax1.set_xlim(10.0,200.0)
ax1.tick_params(labelsize=16)
plt.show()    
```

### Model comparison example: is that an emission line?

Looking at our plot above, there is a hint of a possible feature at around 70 GeV.  Could there be an emission feature there?  This is an example of a hypothesis test with a nested model where the main model is a power-law plus a Gaussian emission feature.  The simple power-law model is nested in the power-law plus Gaussian: it is this model with a single constraint, namely that the Gaussian flux is zero.  Since zero gives a lower bound of the parameter for an emission feature (which is not formally allowed for likelihood ratio tests), we should be sure to allow the Gaussian flux also to go negative (to approximate an absorption feature).  First, let's define the model and our likelihood function (in this case, weighted least squares).  For plotting purposes, we will also define a separate function which outputs the model values as well as a function to make the plot:

```python
def ChiSq_Integ(parm, my_model, ebins, yval, dy): # the weighted least-squares for an array of upper and lower bin 
    # edges (xval1 and xval2) fitted with the integral over the bin of the generic model my_model
    i = 0
    ymod = np.zeros(len(yval))
    for energy in ebins[:-1]:
        ymod[i], ymoderr = scipy.integrate.quad(lambda x: my_model(x, *parm),ebins[i],ebins[i+1])
        ymod[i] = ymod[i]/(ebins[i+1]-ebins[i])  # we now divide by the bin width to match the counts density
        # units of our data
        i=i+1

    return sum(pow((yval-ymod)/dy,2))

def plgauss_model(value, pl_norm, pl_index, line_en, line_sig, line_norm):
    # The line is a Gaussian shape with a normalisation equal to the number of counts in the line
    lineflux = np.exp(-0.5*np.power((value - line_en)/line_sig, 2.))/(line_sig*np.sqrt(2.*np.pi))
    return pl_norm * np.power(value,pl_index) + line_norm * lineflux

def integrate_model(parm, my_model, ebins): 
    #  If we want to be really 'correct'
    ## with our plotting, we should ensure the steps match the bin edges - there is no simple way that we have 
    ## found to do this in Python.  One solution would be not to plot model points at the bin centres 
    ## but to output 2 model values at each bin edge (one equal to the model value in the bin below the edge, 
    ## the other equal to the model value in the bin below the edge). 
    ## The correct steps could then be produced simply by drawing a connecting line through all these points.
    ## Here we implement this approach and produce the correct-format model energies and model values for plotting
    ## a "bespoke" stepped model function.
    i = 0
    j = 0
    xmod = np.zeros(2*len(ebins)-2)
    ymod = np.zeros(2*len(ebins)-2)
    for energy in ebins[:-1]:
        mod, moderr = scipy.integrate.quad(lambda x: my_model(x, *parm),ebins[i],ebins[i+1])
        mod = mod/(ebins[i+1]-ebins[i])  # we now divide by the bin width to match the counts density
        # units of our data
        xmod[j] = ebins[i]
        ymod[j] = mod
        j = j+1
        xmod[j] = ebins[i+1]
        ymod[j] = mod
        j = j+1            
        i=i+1

    return xmod, ymod

def plot_spec(energies, bwidths, cdens, cdens_err, modflux, modens):
    fig, ax1 = plt.subplots(1,1, figsize=(6,6))
    ax1.errorbar(energies, cdens, xerr=bwidths/2., yerr=cdens_err, fmt='o')
    ax1.plot(modens, modflux, color='red')
    ax1.set_xlabel("Energy (GeV)", fontsize=20)
    ax1.set_ylabel("Counts/GeV", fontsize=20)
    ax1.set_yscale('log')
    ax1.set_xscale('log')
    ax1.set_ylim(0.3,300.0)
    ax1.set_xlim(10.0,200.0)
    ax1.tick_params(labelsize=16)
    plt.show()    
```

Now we will look at the problem using `minimize` with the L-BFGS-B method so that we can use bounds to constrain our parameters.  First, we will look at the case where the line has zero flux, i.e. zero counts from the line (we also freeze the other line parameters).  Then we will allow the line flux to vary but keep the line energy fixed so that it doesn't run away from where we think the line is.  We will also keep the line width fixed.  Finally, we will free the line energy and width to try to get the best possible fit.

```python
p0 = [100.0, -1.6, 70.0, 10.0, 0.0]

result =  op.minimize(ChiSq_Integ, p0, args=(plgauss_model,edges2,cdens,cdens_err), method='L-BFGS-B', 
                     bounds=[(None,None),(None,None),(70.0,70.0),(5.0,5.0),(0.0,0.0)])
print result.x # best-fitting parameter values
print result.fun # value of the function minimised (e.g. chi-squared)

modens, modflux = integrate_model(result.x, plgauss_model, edges2)
plot_spec(energies, bwidths, cdens, cdens_err, modflux, modens)

result =  op.minimize(ChiSq_Integ, result.x, args=(plgauss_model,edges2,cdens,cdens_err), method='L-BFGS-B', 
                     bounds=[(None,None),(None,None),(70.0,70.0),(2.0,2.0),(None,None)])
print result.x # best-fitting parameter values
print result.fun # value of the function minimised (e.g. chi-squared)

modens, modflux = integrate_model(result.x, plgauss_model, edges2)
plot_spec(energies, bwidths, cdens, cdens_err, modflux, modens)

result =  op.minimize(ChiSq_Integ, result.x, args=(plgauss_model,edges2,cdens,cdens_err), method='L-BFGS-B', 
                     bounds=[(None,None),(None,None),(None,None),(None,None),(None,None)])
print result.x # best-fitting parameter values
print result.fun # value of the function minimised (e.g. chi-squared)

modens, modflux = integrate_model(result.x, plgauss_model, edges2)
plot_spec(energies, bwidths, cdens, cdens_err, modflux, modens)
```

It's useful to note here an important distinction, between goodness-of-fit (determined from the best-fitting chi-squared value) and the separate question of whether one model is better than another (a hypothesis test).  In the example above, we see that all the different fits are technically 'good'.  E.g. there are 22 data bins and so in the first fit with 20 d.o.f. (since there are 2 free parameters) for a chi-squared value of 20.65 we have a goodness-of-fit of $\sim0.42$, perfectly acceptable!  But a good fit doesn't mean that the model is the best one possible: it just means it is the best fit for that particular model and constrained parameters.  Hypothesis testing using likelihood ratio (or equivalently with Wilks' theorem, $\Delta \chi^{2}$) is the approach to use to determine whether to stop fitting or not.  This does not mean you have to keep fitting and try anything: you should let the data and your physical understanding guide you.  A useful rule-of-thumb is that, if a fit is good and does not show any interesting residuals (at least, ones you cannot attribute to known systematics), you should probably stop there, rather than trying to improve the fit.

In our example above, we can see that the fit has improved by allowing the line flux to be non-zero, with the chi-squared dropping from 20.65 to 14.5, i.e. $\Delta \chi^{2}$, measured from subtracting the worse (higher) value from the better one, is 6.15.  Is this a significant improvement?

One important question here is: what is the number of additional free parameters?  Wilks' theorem tells us that the $\Delta \chi^{2}$ in going from a more constrained model to a less constrained one is itself distributed as $\chi^{2}_{k}$ where $k$ is the number of additional constraints in the more constrained model.  In our case, it seems like $k=3$, but for an emission line we should be careful: the line energy is not really a 'nestable' model parameter.  The line width might be considered as a parameter, but often is limited by the resolution of the instrument which applies a significant lower bound, also making the likelihood ratio approach unsuitable.  Therefore for simplicity here it is better to do the test assuming only the flux as the additional constraint, i.e. the null hypothesis is for flux = 0.  

Thus, we have a $\chi^{2}_{1}$ distribution and we can estimate the significance of our improvement as follows:

```python
chi2 = scipy.stats.chi2(1)
print "p-value for our delta-chi-squared: ",chi2.sf(6.15)
print len(cdens)
```

However, we also need to take into account the fact that (without searching at an already-known fixed energy)  we have done a blind search for the line.  Effectively, we have 'by eye' done $N$ trials at different energies before finding our feature at $\sim 70$~GeV.  Here we can assume that the number of trials corresponds to the number of bins, since this sets a lower limit on the energy resolution.  If the line width depends on energy, e.g. due to the known properties of the instrument, you can do this in a more sophisticated way. Once we have an estimate for the number of trials, we can apply the Bonferroni correction (see the week 4 lecture slides).  A more rigorous Bayesian approach would be to use priors correctly, but that is beyond the scope of this course.

```python
pval = chi2.sf(6.15)

print "p-value corrected for number of trials: ", 1.0-np.power((1.0-pval),len(cdens)) 
```

Which is not very impressive!  Thus we conclude that the apparent emission feature at ~70 GeV is not significant.

### Setting upper limits

The hypothetical GeV photon data used above does not seem to show any clear emission features, but in some cases it is also interesting to consider what limits we can place on any features that are present in the 'true' photon spectrum which our photons are sampled from.  This may be the case if we have a physical model that predicts an emission feature with a given strength (so we can test our model: can we rule it out at some level with our non-detection of the line).  Alternatively, the model may predict the line strength based on one or more of its parameters, and we might want to set limits on the line flux in order to constrain the range of allowed values for those parameters.  Or, in an astronomical setting, we could imagine that a certain type of astronomical source should show emission lines with a particular flux value.  Can we rule out this source type at a given level by setting an upper limit?

Here we will consider a hypothetical example: imagine that we have a theory which predicts that dark matter is a particle which undergoes a decay to produce a GeV photon emission feature at 50 GeV.  Our gamma-ray observatory obtains the spectrum which we loaded in above.  Let's further assume that the line is intrinsically narrow and will therefore appear in the spectrum at the detector's resolution, which we assume to correspond to a 1-sigma width of 2.0 GeV.

To find an upper-limit we need to decide on a significance.  Perhaps we want to find out the 3-sigma constraint on the line flux predicted by our model, so we will use that to set the p-value for our upper limit.  Line energy is fixed by our model, and the line width by the detector, so we just need to step through a range of flux values until we find the one which is just inconsistent with our p-value.  Assuming we step through the line flux with a step-size which is small enough, we can then use this line flux as our upper limit (to be more accurate, we could also interpolate):

```python
##  First let's fit the simple power-law again as our baseline hypothesis.  We'll already set the line energy and
## width to the right values, and freeze the flux at 0:
p0 = [100.0, -1.6, 50.0, 2.0, 0.0]
result =  op.minimize(ChiSq_Integ, p0, args=(plgauss_model,edges2,cdens,cdens_err), method='L-BFGS-B', 
                     bounds=[(None,None),(None,None),(70.0,70.0),(5.0,5.0),(0.0,0.0)])
print result.x # best-fitting parameter values
print result.fun # value of the function minimised (e.g. chi-squared)

p_upper = sigmas_to_p(3.0) ## We want the p-value for the 3-sigma case
chisq_noline = result.fun  ##  We will measure delta-chi-squared with respect to our 0 flux case
delchisq_upper = chi2.isf(p_upper)

fluxvals = []
delchisqvals = []
flag = 0
for flux in np.arange(0.0, 50.0, 1.0):
    ## We need to fit all the free parameters (the power-law continuum parameters), keep line energy and width
    ## fixed, and step through the line flux
    result =  op.minimize(ChiSq_Integ, result.x, args=(plgauss_model,edges2,cdens,cdens_err), method='L-BFGS-B', 
                     bounds=[(None,None),(None,None),(50.0,50.0),(5.0,5.0),(flux,flux)])
    delchisq = result.fun - chisq_noline
    print "Fitting for flux = ", flux, "Delta-chisq = ", delchisq, " with target value ",delchisq_upper
    if (flag == 0 and delchisq >= delchisq_upper):
        print "Delta chi-squared = " + str(delchisq) + " and exceeds upper limit of " + str(delchisq_upper) + \
        " for a flux of " + str(flux) + " counts."
        flag = 1

    fluxvals = np.append(fluxvals,flux)
    delchisqvals = np.append(delchisqvals,delchisq)

## We can also plot our delta-chisq values on a graph (along with a line showing our upper limit value)

fig, ax1 = plt.subplots(1,1, figsize=(6,6))
ax1.plot(fluxvals, delchisqvals)
ax1.set_xlabel("Flux (counts)", fontsize=20)
ax1.set_ylabel("delta-chisq", fontsize=20)
ax1.tick_params(labelsize=16)
ax1.axhline(delchisq_upper, color='r', linestyle='dotted', lw=2)
plt.show()    
```

Thus we have a 3-sigma upper limit of between 34 and 35 counts in our predicted line (to be conservative we can quote 35 counts, but we could also calculate using interpolation or a finer grid between these values), which we can use to limit our underlying physical model parameters.  For interest, we can plot the model with this limiting line flux to see how it compares with the data:

```python
p0 = [100.0, -1.6, 50.0, 2.0, 0.0]
result =  op.minimize(ChiSq_Integ, p0, args=(plgauss_model,edges2,cdens,cdens_err), method='L-BFGS-B', 
                     bounds=[(None,None),(None,None),(50.0,50.0),(2.0,2.0),(35.0,35.0)])

modens, modflux = integrate_model(result.x, plgauss_model, edges2)
plot_spec(energies, bwidths, cdens, cdens_err, modflux, modens)
```

### Dealing with low counts per bin

It is important to remember that Wilks' theorem <i>does not require the data to be normally distributed</i>.  The main requirement is that the model parameters being constrained in the null hypothesis have MLEs which are normally distributed.  This is actually quite a relaxed requirement, which can be satisfied by many data sets and models fitted to them.  In particular, it is useful to bear this point in mind when you wish to do hypothesis testing and obtain confidence intervals or upper limits for data with low counts per bin (so the errors on each bin are not normally distributed).  So, the methods described above may also be applied to data in the low counts limit - the most important difference is that instead of calculating the $\Delta \chi^{2}$ directly from the chi-squared values of each model, one must instead work it out using twice the natural logarithm of the likelihood ratio (e.g. the latter being calculated using Poisson pdfs), as described in the Week 6 lecture slides.
